```python
def update_plot(hidden_layer_size):
    
    # Create Mult_layer Perceptron Classfier
    clf = MLPClassifier(hidden_layer_sizes = (hidden_layer_size,), 
                      activation = 'relu', max_iter=3000, random_state=1) # Number of neurons = Hidden_layer_size

    # Relu help create non_linear boundaries
    
    # Fit the classifier to the data
    clf.fit(X,y)

    # Create a grid of points for visualization
    #These are ID arrays of 100 values each, representing the X and Y coordinates of the grid
    # These are new data points needed to cover the set of data points used for training
    x_vals = np.linspace(X[:,0].min() - 0.1 , X[:,0].max() + 0.1, 100)
    y_vals = np.linspace(X[:,0].min() - 0.1 , X[:,0].max() + 0.1, 100)

    #The resulting both X-plane and Y_plane are 100x100 arrays
    # representing a grid of 10,000 points
    X_plane, Y_plane = np.meshgrid(x_vals, y_vals)

    # grid plane is single 2D array of shape (10000,2)
    # where row represnts a point in the grid
    # Combining the Grids to get our new data points we will use for predicting labels
    grid_points = np.column_stack((X_plane.ravel(), Y_plane.ravel()))

    # predict the class labels for the grid points (for decision boundary)
    Z = clf.predict(grid_points)

    # Z.reshape(X_plane.shape) reshapes Z into a 100x100 array
    # Predicted labels are reshaped so it can be used f
    Z = Z.reshape(X_plane.shape)

    # Predict class labels for original Data Points
    y_pred = clf.predict(X)


    # Plot the decision boundary 
    # it is often used to visualize the decision boundary by plotting
    # the predicted class probabilities of labels across a grid of points
    plt.contour(X_plane, Y_plane, Z, levels=(-0.5,0.5,1.5), cmap= plt.cm.RdYlGn, alpha=0.6)

    # plot thte original data points with their predicted labels
    # Seperate points for each predicted class
    class_0 = y_pred == 0 # Indices of the points predicted as class 0
    class_1 = y_pred == 1 # indices of points predicted as Class 1

    plt.scatter(X[class_0, 0], X[class_0, 1], c="red", edgecolors="k", marker = "o", s=50, label="Predicted Class 0")
    plt.scatter(X[class_1, 0], X[class_1, 1], c="green", edgecolors="k", marker = "o", s=50, label="Predicted Class 1")

    # Adding labels and title
    plt.ylabel('Feature 1')
    plt.xlabel('Feature 2')
    plt.title("Decision Boundary and predicted labels ")
    plt.legend()
    plt.show()
```

```python
# Create a Slider for the Hidden Layer Size
hidden_layer_size_slider = widgets.IntSlider(value=1, min=1,max=10,step=1,description="Hidden Layer Size")
```

```python
# Create a Interactive widget
interactive_plot = interactive(update_plot, hidden_layer_size=hidden_layer_size_slider)
```

```python
# Display the widgets
display(interactive_plot)
```
